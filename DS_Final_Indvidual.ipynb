```python
import pandas as pd
import numpy as np

data = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv")
data2 = data.dropna()
data2['User_Score'] = data2['User_Score'].astype(float)
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
model
x = data2['Critic_Score'].values
y = data2['User_Score'].values
X = x[:, np.newaxis]
model.fit(X, y)
model.score(X,y)#R^2
yfit = model.predict(X)
plt.scatter(x, y, c='k')
plt.plot(X, yfit, c='b');
```
