```python
import pandas as pd

data = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv")
data2 = data.dropna()
data2['User_Score'] = data2['User_Score'].astype(float)
```

```python
#Our Questions:
    #How many games in each genre?
        
    
    #Does the User_Score correlate with the Critic_Score?

        
    
    #What is the breakdown of games sold on each Platform every year?
    
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
  #How many games in each genre?
with sns.axes_style('white'):
    g = sns.catplot( data=data2,x="Genre" ,aspect=2,
                       kind="count", color='steelblue')
    #g.set_xticklabels(step=2)
```

```python
  #Does the User_Score correlate with the Critic_Score?
plt.plot(data2['Critic_Score'],data2['User_Score'], 'o')
```

```python
   #What is the breakdown of games sold on each Platform every year?
data3 = data2.groupby(['Year_of_Release','Platform'])['Global_Sales'].sum()
data3.unstack().plot.area(color=['black','red','blue','yellow','green','orange','cyan','peru','limegreen','grey',
                                'pink','maroon','royalblue','purple','violet','gold','dodgerblue'])
```
